1.1. Скачать следующие данные: kc-house-data и laptop_price

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('kc_house_data.csv', sep=',')
df.head()
```

1.2. Изучите стоимости недвижимости

```python
plt.hist(df['price'], bins=70)
plt.xlabel('Стоимость объекта недвижимости')
plt.ylabel('Количество объектов недвижимости')
plt.title('Стоимость недвижимости');
plt.grid();
```

**Вывод:**
Максимальное количество квартир продается в диапазоне от 280 000 у.е. до 420 000 у.е. Основная доля рынка недвижимости приходится на квартиры стоимостью до 1 000 000 у.е.

1.3. Изучите распределение квадратуры жилой

```python
sns.histplot(df['sqft_living'], bins = 70)
plt.xlabel('Жилые квадратные метры')
plt.ylabel('Количество объектов недвижимости')
plt.title('Распределение квадратуры жилой');
plt.grid();
```

**Вывод:**
Больше всего спросом пользуется недвижимость с жилой площадью от 1000 до 2800 кв.м. Начиная с показателя 2000 кв.м., количество объектов обратно пропорционально жилой площади.

1.4. Изучите распределение года постройки

```python
plt.hist(df['yr_built'], bins = 70)
plt.xlabel('Годы')
plt.ylabel('Количество объектов недвижимости')
plt.title('Распределение года постройки');
plt.grid();
```

**Вывод:**
Пик строительства наблюдается в 2004-2006 гг.,  спад новостроек пришелся на период 1930-1940 гг.

**2 Задача**

2.1. Изучите распределение домов от наличия вида на набережную
* Постройте график
* Сделайте выводы

```python
data_view = df['waterfront'].value_counts()
values = data_view.values
plt.pie(values, labels=['не набережная', 'набережная'],autopct='%.1f%%');
```

**Вывод:**
Количество объектов с видом на набережную составляют менее одного процента на рынке недвижимости.

2.2. Изучите распределение этажей домов

```python
data_floors = df['floors'].value_counts()
names = data_floors.index
values = data_floors.values
plt.figure(figsize=(8, 6))
plt.bar(names, values)
plt.xticks(rotation=100);
plt.title('Этажность')
plt.xlabel('Этажи')
plt.ylabel('Количество домов')
```

**Вывод:**

Основное количество домов имеют этажность 1 и 2. Дома с этажностью 1.5 , 2.5 и 3 занимают малую долю рынка.

2.3. Изучите распределение состояния домов

```python
data_cond = df['condition'].value_counts()
values = data_cond.values
names = data_cond.index
plt.pie(values, labels=names,autopct='%.1f%%');
```

**Вывод:**

Примерно 65% всех обЪектов недвижимости находится в удовлетворительном состояние. Треть жилья в хорошем или отличном состоянии. Около одного процента недвижимости находится в плачевном состоянии.
